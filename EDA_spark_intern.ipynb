```python
# Now building another column "Overall" for better analysis
# Overall column will be categorical column showing gain or loss depending on profit values.
data['overall']=data['Profit'].apply(lambda x: 'gain' if x>0 else 'loss')
```

```python
data.head(10)
```

# 8. Statistical analysis for categorical data types, Chi-square Test is performed.
###  We will perform Chi-Square test to show dependency of other Categorical columns on "Overall" column.

```python
cat_cols = list(data.describe(include = "O").columns)
print(cat_cols)
```

```python
chi_stat=[]
p_value=[]
for i in cat_cols:
    chi_res=chi2_contingency(np.array(pd.crosstab(data[i],data['overall'])))
    chi_stat.append(chi_res[0])
    p_value.append(chi_res[1])
chi_square=pd.DataFrame([chi_stat,p_value])
chi_square=chi_square.T
col=['Chi Square Value','P-Value']
chi_square.columns=col
chi_square.index=cat_cols
```

```python
chi_square
```

```python
chi_square[chi_square["P-Value"]<0.05]
```

```python
features_p = list(chi_square[chi_square["P-Value"]<0.05].index)
print("Significant categorical Features:\n",features_p)
```

## Key Inference from 8:
1. Categorical feature with P-value<0.05, i.e. {'Ship Mode', 'City', 'State', 'Region', 'Category', 'Sub-Category'} are dependent and have correlation with Overall gain and loss.

# 9. Statistical analysis for numerical data types , ANOVA Test is performed
###  We will perform ANOVA test to show dependency of other Numerical columns on "Overall" column.

```python
num_cols = list(data.describe(include = ('int64','float64')).columns)
print(num_cols)
```

```python
f_stat=[]
p_val=[]
for i in num_cols:
    atr_0=data[data['overall']=="loss"][i]
    atr_1=data[data['overall']=="gain"][i]
    a=f_oneway(atr_0,atr_1)
    f_stat.append(a[0])
    p_val.append(a[1])
anova=pd.DataFrame([f_stat,p_val])
anova=anova.T
cols=['F-STAT','P-VALUE']
anova.columns=cols
anova.index=num_cols
```

```python
anova
```

```python
anova[anova["P-VALUE"]<0.05]
```

```python
features_p_n = list(anova[anova["P-VALUE"]<0.05].index)
print("Significant numerical Features:\n",features_p_n)
```

## Key Inference from 9:
1. Numerical feature with P-value<0.05, i.e. {'Sales', 'Discount', 'Profit'} are dependent and have correlation with Overall gain and loss.
